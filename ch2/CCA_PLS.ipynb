```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

 load Palmer Penguin datasets and do a a bit bit of cleanup by removing any rows with missing values

```python
link2data = "data/penguins.csv"
df = pd.read_csv(link2data)
df =df.dropna()
df.head()
```

Since Canonical Correlation Analysis involve two high dimensional data set sets, we will split the Penguinâ€™s data into two datasets X and Y. The dataset X has two variables corresponding to bill length and depth.

```python
X = df[['bill_length_mm','bill_depth_mm']]
X.head()
```

We also need to standardize the variables by subtracting with mean and dividing by standard deviation.

```python
X_mc = (X-X.mean())/(X.std())
X_mc.head()
```

The second dataset Y contains flipper length and body weight. Standardize Y by subtracting with mean and dividing by standard deviation.

```python
Y = df[['flipper_length_mm','body_mass_g']]
Y.head()
Y_mc = (Y-Y.mean())/(Y.std())
Y_mc.head()
```

```python
from sklearn.cross_decomposition import CCA

ca = CCA()
ca.fit(X_mc, Y_mc)
X_c, Y_c = ca.transform(X_mc, Y_mc)
```

The result is two canonical correlate matrices.

```python
print(X_c.shape)
print(Y_c.shape)
```

Understanding the results from Canonical Correlation Analysis
Let us dig deeper to understand the results from canonical correlation analysis and get an intuition. At first, we will use the pair of canonical covariates and see how are they correlated to each other.

Let us add the results from CCA with the meta data corresponding to Penguins data in to a Pandas dataframe.

```python
cc_res = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCY_1":Y_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "CCY_2":Y_c[:, 1],
                       "Species":df.species.tolist(),
                      "Island":df.island.tolist(),
                      "sex":df.sex.tolist()})
cc_res.head()
```

Let us check how correlated the the first pair of canonical covariates are. We use NumPyâ€™s corrcoef() function to compute the correlation. And we can see that the first of pair of canonical covariates are highly correlated.

```python
import numpy as np
np.corrcoef(X_c[:, 0], Y_c[:, 0])
```

We can also compute the correlation between the second pair of covariates and we see that the correlation is not that high.

```python
np.corrcoef(X_c[:, 1], Y_c[:, 1])
```

To further understand the relationship between the pairs of canonical covariates, let us make scatter plot with the first pair of canonical covariates. We can see the high correlation between the first pair of canonical covariates.

```python
from matplotlib import pyplot as plt
sns.set() # Setting seaborn as default style even if use only matplotlib

sns.set_context("talk", font_scale=1.2)
plt.figure(figsize=(10,8))
sns.scatterplot(x="CCX_1",
                y="CCY_1", 
                data=cc_res)
plt.title('Comp. 1, corr = %.2f' %
         np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
```

rom the correlations between the first pair of canonical covariates and species variable in the data set, we can see that our Canonical Correlation Analysis has captured the shared variation between the two datasets. In this example the shared variation or the latent variable behind the first pair of canonical covariates is the species variable.

By carefully looking at the association between CCs and a variable here the latent variable, we can definitely understand the results from our CCA results. Let us take a step further and make a heatmap of correlations between canonical covariates from each of the dataset against our input dataset including the meta data associated with it,

Let us repeat ourselves by creating dataframe with original data and the canonical covariates from the first dataset X. To compute correlation, we convert the character variables into categorical variables and convert them to 0/1/2 codes.

With Pandasâ€™ corr() function we can compute correlation of all variables in the dataframe.

```python
ccX_df = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "Species":df.species.astype('category').cat.codes,
                      "Island":df.island.astype('category').cat.codes,
                      "sex":df.sex.astype('category').cat.codes,
                       "bill_length":X_mc.bill_length_mm,
                      "bill_depth":X_mc.bill_depth_mm})

corr_X_df= ccX_df.corr(method='pearson') 
corr_X_df.head()
```

```python

# second pair of canonical covariates with the dataset 
ccY_df = pd.DataFrame({"CCY_1":Y_c[:, 0],
                       "CCY_2":Y_c[:, 1],
                       "Species":df.species.astype('category').cat.codes,
                      "Island":df.island.astype('category').cat.codes,
                      "sex":df.sex.astype('category').cat.codes,
                       "flipper_length":Y_mc.flipper_length_mm,
                      "body_mass":Y_mc.body_mass_g})
 
# compute correlation with Pandas corr()
corr_Y_df= ccY_df.corr(method='pearson') 
```

Let us make a heatmap with lower triangular correlation matrix. To do that we subset the original correlational matrix with Numpyâ€™s tril() function.

We can make make lower triangular correlation heatmap using Seabornâ€™s heatmap function.

We see a similar pattern as we saw with the correlation map of canonical covariates from dataset X. One of the highlights is as we saw before the first canonical covariate from dataset Y is highly correlated with Species variable. And the second canonical covariate from dataset is Y is correlated with sex variable, suggesting that we can capture the effect of two distinct hidden variables using canonical correlation analysis.

```python
fig, axes = plt.subplots(1, 2, figsize=(25, 8), sharey=True)


axes[0].set_title("Correlation Heatmap of Canonical Correlates from Data X and dataset")
X_df_lt = corr_X_df.where(np.tril(np.ones(corr_X_df.shape)).astype(np.bool))

sns.heatmap(X_df_lt, ax=axes[0], cmap="coolwarm",annot=True,fmt='.1g')
plt.tight_layout()
plt.savefig("Heatmap_Canonical_Correlates_from_X_and_data.jpg",
                    format='jpeg',
                    dpi=100)


 
# Get lower triangular correlation matrix
Y_df_lt = corr_Y_df.where(np.tril(np.ones(corr_Y_df.shape)).astype(np.bool))
 
# make a lower triangular correlation heatmap with Seaborn
axes[1].set_title("Correlation Heatmap of Canonical Correlates from Data Y and dataset")
sns.heatmap(Y_df_lt,ax=axes[1], cmap="coolwarm",annot=True,fmt='.1g')
plt.tight_layout()
plt.savefig("Heatmap_Canonical_Correlates_from_Y_and_data.jpg",
                    format='jpeg',
                    dpi=100)
```

By coloring the scatterplots between the first pair of canonical covariates with the species variable and see how the canonical covariates have capture the underlying variable behind our datasets.

## The following applied other cross decomposition methods adopted to the above dataset example, from https://scikit-learn.org/stable/auto_examples/cross_decomposition/plot_compare_cross_decomposition.html#sphx-glr-auto-examples-cross-decomposition-plot-compare-cross-decomposition-py 


```python
from sklearn.cross_decomposition import PLSCanonical

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#split the dataset into training (70%) and testing (30%) sets
X_train,X_test,Y_train,Y_test = train_test_split(X_mc,Y_mc,test_size=0.3,random_state=0) 

plsca = PLSCanonical(n_components=2)
plsca.fit(X_train, Y_train)
X_train_r, Y_train_r = plsca.transform(X_train, Y_train)
X_test_r, Y_test_r = plsca.transform(X_test, Y_test)

plsca_predit = plsca.predict(X_test)
plsca_RMSE = np.sqrt(mean_squared_error(Y_test, plsca_predit))
print ("plsca_RMSE = ", plsca_RMSE)
```

```python
#PLS regression, with univariate response, a.k.a. PLS1
from sklearn.cross_decomposition import PLSRegression

pls1 = PLSRegression(n_components=2)
pls1.fit_transform(X_train, Y_train)

pls1_predit = pls1.predict(X_test)
pls1_RMSE = np.sqrt(mean_squared_error(Y_test, pls1_predit))
print ("pls1_RMSE = ", pls1_RMSE)
```

```python
from sklearn.cross_decomposition import CCA

cca = CCA(n_components=2)
cca.fit(X_train, Y_train)
X_train_r, Y_train_r = cca.transform(X_train, Y_train)
X_test_r, Y_test_r = cca.transform(X_test, Y_test)

cca_predit = pls1.predict(X_test)
cca_RMSE = np.sqrt(mean_squared_error(Y_test, cca_predit))
print ("cca_RMSE = ", cca_RMSE)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = ['PLS CA', 'PLS1', 'CCA']
MSE = [plsca_RMSE, pls1_RMSE, cca_RMSE ]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, MSE, color='green')
plt.xlabel("Regression Methods")
plt.ylabel("RMSE")
plt.title("Comparing Compare cross decomposition methods")

plt.xticks(x_pos, x)

plt.show()
```
