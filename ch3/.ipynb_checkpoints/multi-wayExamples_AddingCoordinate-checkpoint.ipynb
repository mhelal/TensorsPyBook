# multi-wayExamples - Tensorisation - converting matrices columns to coordinates


Mentioned in the 3.1 motivatin and intuition section in Ch3: Geometry & Algebra of Tensors
   

A Video Analysis with Tensor Decomposition in Python example can be found at:

These examples tensorise common datasets in matrix form using a tensorisation step that is hand tailored to each example, with attempts to generalise. Other examples in other Python files, such as for BSS problem from sound files, Face recognition from images

Full tensors, sparse tensors in dictionary data structure, and very simple quantisation of date time series is presented. 


Sparse arrays are implemented here as dictionary of multi-dimensional index, value pairs. Other exisint implementations in tensorly is explained in:
http://tensorly.org/stable/user_guide/sparse_backend.html

and in scipy :
https://docs.scipy.org/doc/scipy/reference/sparse.html

The ML algorithms that are complemtely built using Linear Algebra and Matrix forms, might need to be updates to use non-linear algebra and nd-arrays as implemented here. The dictionary style used here for sparse arrays might not be optimal. More experimentation is required.

Similar to using polynomial features to perform polynomial regression, using sklearn LinearRegression class, we can also do any change of features, or basis, to achieve basis function regression.

More is explored in Python Data Science Handbook.  Essential Tools for Working with Data by Jake VanderPlas “In Depth: Linear Regression” on page 390. The Gaussian Basis Function is explored at the end of this notebook.

Some Indexing routines can be found in https://numpy.org/doc/stable/reference/arrays.indexing.html

```python
import math
import numpy as np

def getAggValue (currentVal, value, aggFunc):
    if not np.isnan(currentVal) and currentVal != 0 and isinstance (currentVal, (np.ndarray, int, float)):
            if aggFunc == "np.mean":
                currentVal = (aggFunc(value) + currentVal) / 2 # weighted average to allow for incremental averaging
            if aggFunc == "np.sum":
                currentVal = aggFunc(value) + currentVal
            if aggFunc == "np.max":
                if aggFunc(value) > currentVal: 
                    currentVal = aggFunc(value) 
            if aggFunc == "np.min":
                if aggFunc(value) < currentVal:
                    currentVal = aggFunc(value) 
    else:
        currentVal = aggFunc(value)
    return currentVal

def tensorisation_Values (Df, components, value="value", aggFunc=np.mean):
    if components >= Df.shape[1]:
        print ("Number of components must be less or equal to number of the columns in the input matrix. Exiting without creating the tensor")
        return
    # Step 1. decide the tensor shape and range of coordinate values from min to max
    minVals = [] # array of coordinates minimum values
    maxVals = [] # array of coordinates maximum values
    tensorShape = [] # from the step size (basis) from min to max on each coordinate, the shape size is added to this array
    for i in range(components):
        minVal = Df.iloc[:,i].min()
        # this will be an index, therefore starting from zero is necessary
        if minVal < 0:
            Df.iloc[:,i]  = Df.iloc[:,i] + minVal 
            minVal = 0
        if minVal > 0:
            Df.iloc[:,i]  = Df.iloc[:,i] - minVal 
            minVal = 0
        Df.iloc[:,i]  = Df.iloc[:,i].apply(np.floor).astype('int')
        minVals.append(minVal)
        # also the max value need to be positive non-zero, because it will be the tensor shape, and this is guranteed by the update above 
        maxVal = Df.iloc[:,i].max()
        maxVals.append(maxVal)
        #print("mode  " + str(i) + " max value =" + str(maxVal) + ", min value = " + str(minVal))
        tensorShape.append(int(maxVal)+1)
        
    print (tensorShape)
    
    # Step 2. create the array
    
    tensorShape = tuple(tensorShape) 
    tensor_array = np.zeros(tensorShape) 
    
    # Step 3. copy the values from the original dataframe to the created array
        
    # we can create a sparse array here using ideas from scipy.sparse:
    
    # from scipy.sparse import csr_matrix
    # then instead of the multidimensional loop below
    # tensor_array = csr_matrix((np.ones_like(i), (i, j)))
    # csr_matrix( (data,(row,col)), shape=tensorShape)
    # we can convert to dense if needed using:
    # csr_matrix( (data,(row,col)), shape=tensorShape ).todense()
    
    '''
    for index, row in Df.iterrows():
        t_index = tuple(
            int(row[i]) for i in range(components)
        )
        # update the values in the array, to be used as indices
        for i in range(components):
            row[i] = int(math.floor(row[i])) + abs(int(minVals[i])) + 1
        # either append to lists
        indices.append(t_index)
        values.append(row[value]) # or as computed below for the different functions
    indices = np.array(indices)
    values = np.array(values) 
    tensor_array = sp.csr_array( (values,(indices)), shape=tensorShape) # at some point this function might accept nd-arrays, it accepts matrices only now
        
    # another possible syntax when nd sparse arrays are possible, so far csr_matrix and csr_array seems to be the same
    tensor_array == np.NAN
    for index, row in Df.iterrows():
        t_index = tuple(
            int(row[i]) for i in range(components)
        )
        # update the values in the array, to be used as indices
        for i in range(components):
            row[i] = int(math.floor(row[i])) + abs(int(minVals[i])) + 1
        if tensor_array == np.NAN: # if this is the first element
            tensor_array = sp.csr_matrix( (aggFunc(row[value]),(t_index)), shape=tensorShape)
        else: # other wise stack together
            tensor_array2 = sp.csr_matrix( (aggFunc(row[value]),(t_index)), shape=tensorShape)
            tensor_array = sp.vstack([tensor_array, tensor_array2])
    
    also we can come up with a new sparse tensor structure such as dictionary of index tuple and aggregated value pairs, but will need to build a library to update
    sklearn and tensor flow and others to use this datastructures for all ML algorithms.
    
    a dictionary is used in the next tensorisation function with conditions and quantisation
    
    '''
    #full dense nd-array
    for index, row in Df.iterrows():
        t_index = tuple(
            int(row[i]) for i in range(components)
        )
        # update the values in the array, to be used as indices
        #for i in range(components):
        #    row[i] = int(math.floor(row[i])) + abs(int(minVals[i])) + 1
        #print (t_index)
        tensor_array[t_index] = getAggValue (tensor_array[t_index], row[value], aggFunc)
        
    return tensor_array
```

```python
# we will need to update the tensorisation function to be able to aggregate using a condition per column in the dataframe 
# by mean or other aggregation function, and take
# the values as mean of rows returned from the condition 
# this is almost tailored to Latitude & Longitude, and later the data specific values, it is better to be updated to ranges, such 
# that values are aggregated when they are >= the current basis index and < next basis index
# many optimisations can be achived for vectorisation, paralleisation, quantisation (make the basis larger such that the dataset size will be smaller)
def tensorisation2_Values (Df, components, value="value", aggFunc=np.mean, sparse = False, quantise = False, qFactor = 0):
    if components >= Df.shape[1]:
        print ("Number of components must be less or equal to number of the columns in the input matrix. Exiting without creating the tensor")
        return
    # Step 1. decide the tensor shape and range of coordinate values from min to max
    minVals = []
    maxVals = []
    tensorShape = []
    for i in range(components):
        minVal = Df.iloc[:,i].min()
        # this will be an index, therefore starting from zero is necessary
        if minVal < 0:
            Df.iloc[:,i]  = Df.iloc[:,i] + minVal 
            minVal = 0
        if minVal > 0:
            Df.iloc[:,i]  = Df.iloc[:,i] - minVal 
            minVal = 0
        Df.iloc[:,i]  = Df.iloc[:,i].apply(np.floor).astype('int')
       
        minVals.append(minVal)
        # also the max value need to be positive non-zero, because it will be the tensor shape, and this is guranteed by the update above 
        maxVal = Df.iloc[:,i].max()
        maxVals.append(maxVal)
        #print("mode  " + str(i) + " max value =" + str(maxVal) + ", min value = " + str(minVal))
        #print("mode  " + str(i) + " max value =" + str(Df.iloc[:,i].max()) + ", min value = " + str(Df.iloc[:,i].min()))
        if not quantise:
            # the basis for this coordinate is step 1
            tensorShape.append(int(maxVal)+1)
        else:
            # the basis for this coordinate is step 1 + qFactor, then the maximum dimension size is int(maxVal) / qFactor
            tensorShape.append((int(maxVal) / qFactor)+1)
    # update the values in the array, to be used as indices
    print (tensorShape)
    tensorShape = tuple(tensorShape) 
    
    # Step 2. create the array
    if not sparse:
        tensor_array = np.zeros(tensorShape)
    else:
        tensor_array = {}

    
    '''
    ## two useful functions:
    
    np.unravel_index(0, tensorShape) # flat linear index to multidimensional index 
    np.ravel_multi_index([tensorShape[i]-1 for i in range(len(tensorShape))], tensorShape) # multidimensional index to flat linear index

    '''
    # Step 3. copy the values from the original dataframe to the created array or sparse dictionary
    indexCols = [Df.columns[i] for i in range(components)] # create index column names
    gdf = Df.groupby(indexCols).apply(lambda x: aggFunc(x[value])) # groupby and aggregate
    for t_index, row in gdf.iteritems(): # iterate through the multi dimensional index tupr and the aggregated values in row
        condition = '' # create the conditions in case more rows are found to aggregate on by conditions, this might not be useful, and might be useful for quantisation
        for j in range(len(t_index)):       # iterate through the dataframe columns
            ## update the values in the array, to be used as indices
            #row[i] = int(math.floor(row[i])) + abs(int(minVals[i])) + 1
            if not quantise:
                condj = '(Df[Df.columns[' + str(j)+']] == ' + str(t_index[j]) + ')'                
            else:
                # quantisation can aggregate a number of bases together such as
                condj = '(Df[Df.columns[' + str(j)+']] >= ' + str(t_index[j]) + ') & (Df[Df.columns[' + str(j)+']] < ' + str(t_index[j] + qFactor) + ')'
            #print ("j: " + str(j) + " column: " + Df.columns[j] + " cond: " + condj)
            if j==0: # the following is without quantisation
                condition += condj
            else:
                condition += ' & ' + condj  
        #print (t_index,  condition)
        
        if not sparse:
             currentVal = tensor_array[t_index] # reading from numpy array
        else:
            if t_index in tensor_array:
                currentVal = tensor_array[t_index] # reading from sparse dictionary
            else:
                currentVal = 0
                
        if len(condition) > 0:    
            #print (eval(condition))
            aggValue = getAggValue (currentVal, Df.loc[eval(condition)][value], aggFunc)
        else:
            aggValue = getAggValue (currentVal, Df.loc[value], aggFunc)
        
        if not sparse:
            # this is for full array
            tensor_array[t_index] = aggValue
        else:
            # this for sparse dictionary
            tensor_array.update({t_index:aggValue})

        #print (t_index,  condition, currentVal, aggValue, tensor_array[t_index])  
    
   
    # the following works for small tensors, and take long or even impossible for large tensors
    
    '''
    for i in range (np.prod(tensorShape)-1): # iterate through the tensor flat indices, can try for element in np.nditer(arr): then get the index of the element
        t_index = np.unravel_index(i, tensorShape) # get the tensor multidimensional index
        #print ("i: " + str(i) + " t_in: " + str(t_index))
        condition = '' # accumulate the conditions to add to the data frame selection
        for j in range(len(t_index)):       # iterate through the dataframe columns
            if j==0:
                condition += '(Df[Df.columns[' + str(j)+']] == ' + str(t_index[j]) + ')'
            else:
                condition += ' & (Df[Df.columns[' + str(j)+']]  == ' + str(t_index[j]) + ')'
            #print ("j: " + str(j) + " column: " + Df.columns[j] + " cond: " + condition)
            
        #print ("Condition is: " + condition)
        
        # this is for full array
        if len(condition) > 0:    
            #print (eval(condition))
            tensor_array[t_index] = aggFunc(Df.loc[eval(condition)][value])
        else:
            tensor_array[t_index] = aggFunc(Df.loc[value])
        # for sparse arrays we can insert a dictionary of index tuple and aggregated value using the quantisation information
        # for example, if the basis is one value change in the coordinate of the latitude and longtitude, we can use
        # qfactor = 10, to make the aggregation from 0 to 10 in one value on one coordinate value of 0, and from 11 to 20 is 
        # aggregated on one coordinate value of 1, and so forth.
    '''
    return tensor_array
```

### Linear Regression using Gaussian Basis Functions
##### Python Data Science Handbook.  Essential Tools for Working with Data by Jake VanderPlas 

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.base import BaseEstimator, TransformerMixin

class GaussianFeatures(BaseEstimator, TransformerMixin):
    """Uniformly-spaced Gaussian Features for 1D input"""
    
    def __init__(self, N, width_factor=2.0):
        self.N = N
        self.width_factor = width_factor
    
    @staticmethod
    def _gauss_basis(x, y, width, axis=None):
        arg = (x - y) / width
        return np.exp(-0.5 * np.sum(arg ** 2, axis))
        
    def fit(self, X, y=None):
        # create N centers spread along the data range
        self.centers_ = np.linspace(X.min(), X.max(), self.N)
        self.width_ = self.width_factor * (self.centers_[1] - self.centers_[0])
        return self
        
    def transform(self, X):
        return self._gauss_basis(X[:, :, np.newaxis], self.centers_,
                                 self.width_, axis=1)

rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)
xfit = np.linspace(0, 10, 1000)

gauss_model = make_pipeline(GaussianFeatures(10, 1.0),
                            LinearRegression())
gauss_model.fit(x[:, np.newaxis], y)
yfit = gauss_model.predict(xfit[:, np.newaxis])

gf = gauss_model.named_steps['gaussianfeatures']
lm = gauss_model.named_steps['linearregression']

fig, ax = plt.subplots()

for i in range(10):
    selector = np.zeros(10)
    selector[i] = 1
    Xfit = gf.transform(xfit[:, None]) * selector
    yfit = lm.predict(Xfit)
    ax.fill_between(xfit, yfit.min(), yfit, color='gray', alpha=0.2)

ax.scatter(x, y)
ax.plot(xfit, gauss_model.predict(xfit[:, np.newaxis]))
ax.set_xlim(0, 10)
ax.set_ylim(yfit.min(), 1.5)

```
